```python
# Create a list of 1s in 15 times using plain python
ones_list = []
for i in range(15):
    ones_list.append(1)
print(ones_list)
```

```python
# List of 15 numbers of 7s
sevens_list = []
for i in range(15):
    sevens_list.append(7)
print(sevens_list) 
```

```python
# List numbers from 100 to 150
numbers_list = []
for i in range(100, 151):
    numbers_list.append(i)
print(numbers_list)
```

```python
# List of even numbers between 0 and 100
even_numbers = []
for i in range(0, 101):
    if i % 2 == 0:
        even_numbers.append(i)
print(even_numbers)
```

```python
# List the years that are leap years between 1950 and 2020 devisible by 4
leap_years = []
for year in range(1950, 2021):
    if year % 4 == 0:
        leap_years.append(year) 
print(leap_years)
```
